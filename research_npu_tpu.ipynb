In this notebook we compare the strengths and limitations of Neural Processing Units (NPUs) and Tensor Processing Units (TPUs) for video analytics via YOLO networks.

Key takeaways:
1. NPU are better for real-time inference in retail analytics, traffic monitoring, industrial safety monitoring.
2. TPU are better for historical batch, large-scale video analytics (YouTube ingestion, city-wide video archives).

The notebook is structured as follows. In section [1](#npus-and-tpus-comparison-table) we compare NPUs and TPUs in terms of computations performance, monetary costs.

## NPUs and TPUs (Comparison Table)

NPUs are more suitable in the following cases:
- for real-time inference, video analytics with regions of interest (ROI) (retail analytics, traffic monitoring, industrial event logging for safety)
- event logging with temporal filters, cooldowns, multi-threading, multi-stream YOLO inference.

TPU are more suitable in the following cases: 
- large-scale batch processing (citywide traffic monitoring, YouTube ingestion, smart city video archives).
- cloud-based video search, historical batch analysis, city-wide video archives.

<div class="alert alert-block alert-info">

A <b>Neural Processing Unit</b> (NPU) is a specialized hardware accelerator to run neural network computations (matrix multiplications, tensor operations, convolutions). 

A <b>Tensor Processing Unit </b> (TPU) is a specialized hardware accelerator to run neural network computations in Google's TensorFlow ecosystem.
</div>

NPUs:
- have high performance per watt.
- designed for real-time AI inference, not general computing, training.
- tied to specific software stacks (TensorRT, NNAPI, TFLite).

TPUs:
- optimized for TensorFlow operations (matrix multiplication, convolutions). 
- designed for both model train and inference.
- Cloud TPUs require Google Cloud infrastructure.
- Edge TPUs only support quantized TFLite models.

<b>Examples of NPU:</b> 

- Huawei Ascend NPU is used for traffic monitoting systems in China.

- Qualcomm Snapdragon (Hexagon DSP) is used in retail shelf analytics where we run YOLO on NPU to detect stock levels in real time.
- Apple Neural Engine is used in apps for live detection: pets, vehicles, people tracking.
- Microsoft Azure Percept NPU (Vision AI Dev Kit) are embedde in edge AI cameras running YOLO for people counting, zone entry/exit logging, vehicle detection, used in smart retail (customer flow tracking) and factory monitoring.

<b>Examples of TPU:</b> 

- YouTube ingestion
- BERT, GPT
