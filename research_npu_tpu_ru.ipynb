В этом блокноте мы сравниваем сильные стороны и ограничения нейронных процессоров (NPU) и тензорных процессоров (TPU) для видеоаналитики с использованием сетей YOLO.

Основные выводы:
1. NPU лучше подходят для инференса в реальном времени в аналитике розничной торговли, мониторинге трафика и промышленном мониторинге безопасности.
2. TPU лучше подходят для пакетной обработки исторических данных и крупномасштабной видеоаналитики (обработка видео для YouTube, городские видеоархивы).

Блокнот структурирован следующим образом. В разделе [1](#npus-and-tpus-comparison-table) мы сравниваем NPU и TPU по производительности вычислений и денежным затратам.

## NPU и TPU (Сравнительная таблица)

NPU более подходят в следующих случаях:
- для инференса в реальном времени, видеоаналитики с областями интереса (ROI) (аналитика розничной торговли, мониторинг трафика, промышленная регистрация событий для безопасности)
- регистрация событий с временными фильтрами, задержками, многопоточностью, многопоточным инференсом YOLO.

TPU более подходят в следующих случаях: 
- крупномасштабная пакетная обработка (общегородской мониторинг трафика, загрузка видео на YouTube, видеоархивы умного города).
- облачный поиск видео, исторический пакетный анализ, общегородские видеоархивы.

<div class="alert alert-block alert-info">

<b>Нейронный процессор</b> (NPU) — это специализированный аппаратный ускоритель для выполнения вычислений нейронных сетей (матричные умножения, тензорные операции, свертки). 

<b>Тензорный процессор</b> (TPU) — это специализированный аппаратный ускоритель для выполнения вычислений нейронных сетей в экосистеме Google TensorFlow.
</div>

NPU:
- обладают высокой производительностью на ватт.
- разработаны для ИИ-инференса в реальном времени, а не для вычислений общего назначения или обучения.
- привязаны к конкретным программным стекам (TensorRT, NNAPI, TFLite).

TPU:
- оптимизированы для операций TensorFlow (матричное умножение, свертки). 
- разработаны как для обучения моделей, так и для инференса.
- Облачные TPU требуют инфраструктуры Google Cloud.
- Edge TPU поддерживают только квантованные модели TFLite.

<div class="alert alert-block alert-success">
<b>Примеры NPU:</b> 

- NPU Huawei Ascend используется в системах мониторинга трафика в Китае.
- NPU Samsung Exynos используется в камерах интеллектуального наблюдения, которые обнаруживают события движения на основе ROI (люди, входящие в запретные зоны).
- Qualcomm Snapdragon (Hexagon DSP) используется в аналитике полок розничных магазинов, где мы запускаем YOLO на NPU для определения уровней запасов в режиме реального времени.
- Apple Neural Engine используется в приложениях для обнаружения в реальном времени: отслеживание домашних животных, транспортных средств, людей.
- NPU Microsoft Azure Percept (Vision AI Dev Kit) встроены в периферийные ИИ-камеры, работающие на YOLO для подсчета людей, регистрации входа/выхода из зон, обнаружения транспортных средств; используются в умной рознице (отслеживание потока покупателей) и мониторинге заводов.

<b>Примеры TPU:</b> 

- Обработка видео для YouTube
- BERT, GPT
